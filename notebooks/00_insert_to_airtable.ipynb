```python
import pyairtable
import pandas as pd
import numpy as np
import os

from pyairtable import Table
from dotenv import load_dotenv
```

```python
# Communityt table access
API_KEY = os.environ.get('AIRTABLE_API')
BASE_ID = os.environ.get('BASE_ID')
TABLE_NAME = os.environ.get('TABLE_NAME')
```
